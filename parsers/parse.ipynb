```python
import numpy as np
```

```python
file = open("../ITU_data.csv", "r")
lines = np.array([line.strip().split(",") for line in file.readlines()[1:]])
file.close()
```

```python
catchments = np.unique(lines[:, 6])
catchments
```

```python
actual_numbers = []

for c in catchments:
  try:
    n = int(c)
    actual_numbers.append(n)
    print(n)

  except Exception as e:
    print(e)
    print(c)
    pass

print(len(actual_numbers))

```
